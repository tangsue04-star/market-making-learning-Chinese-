```python
import ccxt
from datetime import datetime, timedelta

# 拉BTC/USD 历史K线（Coinbase）
exchange = ccxt.coinbase({
    'rateLimit': 1000,
    'enableRateLimit': True,
})

# 拉最近30根小时线（每根1小时）
symbol = 'BTC/USD'
ohlcv = exchange.fetch_ohlcv(symbol, timeframe='1h', limit=30)

# 收盘价列表（从老到新）
close_prices = [candle[4] for candle in ohlcv]

# 算均线
def sma(data, period):
    """简单移动平均线（Simple Moving Average）"""
    if len(data) < period:
        return None
    return sum(data[-period:]) / period

ma5 = sma(close_prices, 5)
print(f"MA5 (最近5小时平均价): ${ma5:.2f}")

ma20 = sma(close_prices, 20)
print(f"MA20 (最近20小时平均价): ${ma20:.2f}")

if ma5 and ma20:
    if ma5 > ma20:
        print(f"信号: MA5 > MA20 → 🟢 短期趋势向上（金叉）")
    else:
        print(f"信号: MA5 < MA20 → 🔴 短期趋势向下（死叉）")
```
