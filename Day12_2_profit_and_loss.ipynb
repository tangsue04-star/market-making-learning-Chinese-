```python
import pandas as pd

# 模拟交易记录
trades = [
    {"side": "buy",  "price": 63500, "qty": 0.5, "fee": 3.18},
    {"side": "sell", "price": 63550, "qty": 0.5, "fee": 3.18},
    {"side": "buy",  "price": 63400, "qty": 1.0, "fee": 6.34},
    {"side": "sell", "price": 63480, "qty": 0.3, "fee": 1.90},
]

current_price = 63500  # 当前市价

def calc_pnl(trades, current_price):
    realized = 0.0
    inventory = 0.0
    cost_basis = 0.0

    for t in trades:
        if t["side"] == "buy":
            inventory += t["qty"]
            cost_basis += t["price"] * t["qty"]
            realized -= t["fee"]  # 手续费算成本
        else:  # sell
            inventory -= t["qty"]
            # 卖出收入
            sell_value = t["price"] * t["qty"]
            # 对应买入成本（用均价）
            avg_cost = cost_basis / inventory if inventory > 0 else t["price"]
            realized += sell_value - (t["price"] * t["qty"])  # 简化版
            realized -= t["fee"]
            cost_basis -= avg_cost * t["qty"]

    # 未实现盈亏（库存按市价重估）
    unrealized = inventory * current_price - cost_basis

    return {
        "inventory": round(inventory, 4),
        "realized_pnl": round(realized, 2),
        "unrealized_pnl": round(unrealized, 2),
        "total_pnl": round(realized + unrealized, 2)
    }

result = calc_pnl(trades, current_price)
print(f"当前库存: {result['inventory']} BTC")
print(f"已实现P&L: ${result['realized_pnl']}")
print(f"未实现P&L: ${result['unrealized_pnl']}")
print(f"总P&L:     ${result['total_pnl']}")
```
