```python
# 🔧 CoinGlass查到的实时BTC资金费率

funding_rates = {
    'Binance': 0.000058,   # 0.0058% → 0.000058
    'OKX':     0.000058,   # 0.0058% → 0.000058
    'Bybit':   -0.000028,  # -0.0028% → -0.000028
}

from datetime import datetime

print(f"{'='*55}")
print(f"  BTC/USDT 资金费率 — CoinGlass手动查表")
print(f"  {datetime.now().strftime('%Y-%m-%d %H:%M')} UTC")
print(f"{'='*55}")

for name, rate in funding_rates.items():
    rate_pct = rate * 100

    if rate > 0.0001:
        sentiment = "🟢 多头偏强" if rate < 0.001 else "🔴 极度看多（过热！）"
    elif rate < -0.0001:
        sentiment = "🔴 空头偏强" if rate > -0.001 else "🟢 极度看空（反转信号）"
    else:
        sentiment = "⚪ 多空均衡（接近0）"

    print(f"\n  📍 {name}")
    print(f"     费率: {rate_pct:.4f}%")
    print(f"     判断: {sentiment}")

print(f"\n{'─'*40}")
print(f"  加权均价: +0.0060%")
print(f"  解读: 资金费率极低，多空接近均衡")
print(f"  BTC刚从$59K反弹回$60K，但多头并没追着加仓")
print(f"  → 市场对反弹持怀疑态度，等周五非农给方向")
print(f"{'='*55}")
```
