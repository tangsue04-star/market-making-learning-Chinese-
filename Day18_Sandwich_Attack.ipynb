```python
import random
import statistics

def simulate_sandwich(
    token_price=100.0,
    user_size=10.0,
    slippage=0.005,
    mev_advance=0.002,
    mev_sell_slip=0.001
):
    max_price = token_price * (1 + slippage)
    mev_buy_price = token_price * (1 + mev_advance)
    mev_buy_cost = mev_buy_price * user_size
    user_fill_price = mev_buy_price * (1 + mev_advance)
    user_cost = user_fill_price * user_size
    mev_sell_price = user_fill_price * (1 - mev_sell_slip)
    mev_revenue = mev_sell_price * user_size
    mev_profit = mev_revenue - mev_buy_cost
    user_extra = user_cost - (token_price * user_size)
    return {
        "当前价": token_price, "用户购买量": user_size,
        "MEV买入价": round(mev_buy_price, 2),
        "用户成交价": round(user_fill_price, 2),
        "MEV卖出价": round(mev_sell_price, 2),
        "MEV利润": round(mev_profit, 2),
        "用户多付": round(user_extra, 2),
        "用户总成本": round(user_cost, 2)
    }

def batch_simulation(trials=1000):
    results = []
    for _ in range(trials):
        price = random.uniform(10, 500)
        size = random.uniform(1, 50)
        slip = random.uniform(0.001, 0.02)
        adv = random.uniform(0.0005, 0.005) #mev_advance=adv
        sell_slip = random.uniform(0.0005, 0.003)
        r = simulate_sandwich(
            token_price=price, user_size=size,
            slippage=slip, mev_advance=adv,
            mev_sell_slip=sell_slip
        )
        results.append(r["MEV利润"])
    print("模拟次数:", trials)
    print("MEV利润均值:", round(statistics.mean(results), 2))
    print("MEV利润中位数:", round(statistics.median(results), 2))
    print("MEV利润最大:", round(max(results), 2))
    print("MEV利润最小:", round(min(results), 2))

r = simulate_sandwich()
for k, v in r.items():
    print(k + ":", v)
print()
batch_simulation(1000)
```
